```python
import statsmodels.formula.api as smf
model1 = smf.ols("sunday~daily",data1).fit()
```

```python
model.summary()
```

#### observation
- the predicted equation is (y_hat = beta_node +beta_one x )expressed like that from the result
- beta_node = 13.8356
- beta_one = 1.3397

- the probability(p-value) for intercept (beta_0) is 0.707 > 0.05
- therefore the intercept coefficient may not be that much significant in prediction
- hpwever the p-value for "daily" (beta_1) is 0.00 < 0.05
- therfore the beta_1 coefficent is highly significant and is contributint to predicition

#### predict for new data

```python
newdata=pd.Series([200,300,1500])
```

```python
data_pred=pd.DataFrame(newdata,columns=['daily'])
data_pred
```

```python
model1.predict(data_pred)
```

```python
pred = model1.predict(data1["daily"])
pred
```

```python
# add predicted valuesa as a column in data1
data1["Y_hat"] = pred
data1
```

```python
#compute the erroe values (residuals) and add as another column
data1["residuals"] = data1["sunday"] - data1["Y_hat"]
data1
```

```python
# compute Mean squared error for the model

mse = np.mean((data1["daily"]-data1["Y_hat"])**2)
rmse = np.sqrt(mse)
print("MSE: ",mse)
print("RMSE: ",rmse)
```

```python
plt.scatter(data1["Y_hat"], data1["residuals"])
```

#### Observation
- the residuals are randomly scattered
hence the assumption of homoscedasticity is satisfied(constant variannce in residuals)

```python
import statsmodels.api as sm
sm.qqplot(data1["residuals"], line='45', fit=True)
plt.show()
```

```python
sns.histplot(data1["residuals"], kde = True)
```

#### observations
- the data points are seen to closely follow the reference line of nomarality
- hence the residuals are approximately normally distributed as also can be seen from the kde plot
- all the above assumptions are verifed
